```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/sierraleone-bumbuna.csv')
```

```python
df
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df = df.set_index('Timestamp')
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna(axis=1)
```

```python
df.loc[df['DHI']<0, 'DHI'] = 0.0
df.loc[df['GHI']<0, 'GHI'] = 0.0
df.loc[df['DNI']<0, 'DNI'] = 0.0
```

```python
df
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outlier = ((df < lower_bound) | (df > upper_bound)).sum()
print('Outlier for each feature')
print(outlier)
```

```python
outlier.plot(kind='bar')
```

```python
#Scatter plots to show the relationships
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.scatterplot(x='RH', y='Tamb', data=df)
plt.title('RH vs. Tamb')
plt.subplot(1, 3, 2)
sns.scatterplot(x='RH', y='GHI', data=df)
plt.title('RH vs. GHI')
plt.subplot(1, 3, 3)
sns.scatterplot(x='RH', y='DNI', data=df)
plt.title('RH vs. DNI')
plt.tight_layout()
plt.show()
```

```python
# Pair plot to visualize pairwise relationships
sns.pairplot(df[['GHI', 'DHI', 'Tamb', 'ModA', 'ModB']])
plt.suptitle('Pair Plot: Solar Radiation and Temperature Measures', y=1.02)
plt.show()
```

```python
cor = df[['GHI', 'DHI', 'Tamb', 'ModA', 'ModB']].corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(cor, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Solar Radiation and Temperature Measures')
plt.show()
```
